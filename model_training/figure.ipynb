```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
confusion_df = pd.DataFrame({
    "Logistic_regression": [116328, 393828, 105554, 295631],
    "MLP": [249174, 368712, 130670, 162785],
    "Truth": [411959, 499382, 0, 0]},
    index=['True_positive', 'True_negative', 'False_positive', 'False_negative'])
```

```python
confusion_df
```

```python
confusion_df.plot(
    kind='bar',
    figsize=(10, 6),
    color=["orange", "green", "grey"]
)

plt.ylabel('Counts')
plt.legend()
plt.grid(alpha=0.2)
plt.xticks(rotation=0)
plt.savefig("confusion_matrix_bar.png", dpi=300, bbox_inches='tight')
plt.show()
```

```python
import numpy as np

labels = confusion_df.index
components = ['True_Positive', 'True_Negative', 'False_Positive', 'False_Negative']
colors = ['green', 'blue', 'orange', 'red']
width = 0.2

x = np.arange(len(labels))  # the label locations

plt.figure(figsize=(10, 6))
for i, (comp, color) in enumerate(zip(components, colors)):
    plt.bar(x + i*width, confusion_df[comp], width=width, label=comp.replace('_', ' '), color=color)

plt.xticks(x + width*1.5, labels)
plt.title('Confusion Matrix Components')
plt.legend(title='Confusion Matrix Components')
plt.show()
```
