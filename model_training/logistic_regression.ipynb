```python
import pandas as pd
import sklearn as skl
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegressionCV
```

```python
data = pd.read_csv('/home/s2106664/msc_project/featurised_data.csv')
data
```

# Train test split using subset datatsets

```python
X_train, X_test, y_train, y_test = train_test_split(data.drop(columns=["sequence_id", "disease_status"]),
                                                    data["disease_status"], train_size=3000000, test_size=1000000,
                                                    random_state=42, shuffle=True)
```

```python
lg = LogisticRegressionCV(cv=10, solver="saga", penalty="l1", max_iter=100,
                          n_jobs=15, scoring="f1", random_state=42)

lg.fit(X_train, y_train)
```
